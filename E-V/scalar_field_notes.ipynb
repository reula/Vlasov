## The wave equation is:

$$
-c ∂^2_t \phi + 2b∂_t \partial_r \phi + a \partial^2_r \phi = -f ∂^2_t \phi + 2(f-1)∂_t \partial_r \phi + (2-f) \partial^2_r \phi = 0

Using $v:= ∂_t \phi$, $w := ∂_r \phi$ it becomes, 

$$
-f \sigma v + 2(f-1)\sigma v + (2-f)w = 0
$$

Together with the equation, 

$$ 
\sigma w = v

$$ 
\left(
