## Eistein's Equations of Spherical Symmetry



### Metric from the paper *On free evolution of self gravitating, spherically symmetric waves.*

$$
ds^2= -g(t,r)(2-f(t,r))dt^2+2(f(t,r)-1)g(t,r)drdt+f(t,r)g(t,r)dr^2+dΩ^2
$$

$$
T_{\mu\nu} = \left(
    \begin{array}{cccc} 
    -g(2-f) & g(f-1) & 0 & 0 \\
    g(f-1) & gf & 0 & 0 \\
    0 & 0 & r^2 sin^2(\theta) & 0 \\
    0 & 0 & 0 & r^2
    \end{array}
    \right)
$$

$$
g^{\mu\nu} = \left(
    \begin{array}{cccc} 
    -f/g & (f-1)/g & 0 & 0 \\
    (f-1)/g & (2-f)/g & 0 & 0 \\
    0 & 0 & r^{-2}/ sin^2(\theta) & 0 \\
    0 & 0 & 0 & r^{-2}
    \end{array}
    \right)
$$

The determinant of the metric is 

$$
\sqrt{-g_{\mu\nu}} = gr^2
$$

The non zero christoffell symbols are (this is for completeness, we are not going to use them here):

\begin{align}
\Gamma^{t}_{tt}={\frac { \left( - f ^{2}+3\,f -2 \right) {\frac {\partial g }{\partial r}} + \left( f  ^{2
}-2\,f +2 \right) {\frac {\partial g}{\partial t}} -g  \left(  \left( f -1 \right) {\frac {\partial f}{\partial r}}  -  {\frac {\partial f}{\partial t}}  \left( f -2 \right)  \right) }{2 g }}
\end{align}

\begin{align}
\Gamma^{r}_{tt}={\frac { \left( f -2 \right) ^{2}{\frac {
\partial g}{\partial r}} + \left( -  f ^{2}+3\,f -2 \right) {
\frac {\partial g}{\partial t}} +g \left(  \left( f -2 \right) {\frac {
\partial f}{\partial r}} -  {\frac {\partial 
f}{\partial t}}   \left( f -3 \right)  \right) }{2g }}
\end{align}

\begin{align}
\Gamma^{t}_{tr}={\frac {g   \left( f -1
 \right) {\frac {\partial f }{\partial t}}  -f  \left(  \left( f -2 \right) {
\frac {\partial g}{\partial r}} + \left( -f  +1 \right) {\frac {\partial g}{\partial t}} +g {\frac {\partial f}{\partial r}}  \right) }{2g}}
\end{align}

\begin{align}
\Gamma^{r}_{tr}={\frac { \left( f  ^{2}-3\,f +2 \right) {\frac {\partial g }{\partial r}} +g \left( f -1 \right) {\frac {\partial f}{\partial r}} - \left( f -2 \right)  \left( {\frac {\partial g }{\partial t}} f  +g  {\frac {\partial f}{\partial t}}  \right) }{2g }}
\end{align}

\begin{align}
\Gamma^{t}_{rr}={\frac {-g  \left( f +1

\end{align}

\Gamma^{r}_{rr}={\frac { \left(  f ^{2}-2\,f +2 \right) {\frac {\partial g}{\partial r}} -g  \left( f -1 \right) {\frac {\partial f}{\partial t}} -f   \left(  \left( f-1\right) {\frac {\partial g}{\partial t}} -g {\frac {\partial f}{\partial r}}  \right) }{ 2g }}
\end{align}


\begin{align}
\Gamma^{t}_{\theta \theta}&= -{\frac { \left( f  -1 \right) r}{g }} \\
\Gamma^{r}_{\theta \theta}&= {\frac { \left( f  - 2 \right) r}{g }} \\
\Gamma^{\theta}_{r \theta}&=\Gamma^{\phi}_{r \phi} = \frac{1}{r}\\
\Gamma^{t}_{\phi \phi}&= \sin\theta^2 \Gamma^{t}_{\theta \theta}\\
\Gamma^{r}_{\phi \phi}&= \sin\theta^2 \Gamma^{r}_{\theta \theta}\\
\Gamma^{\phi}_{\theta \phi}&= \frac{ \cos \theta }{\sin \theta}\\
\Gamma^{\theta}_{\phi \phi}&= - \sin \theta \cos\theta
\end{align}


### Einstein equations including Cosmological constant:

\begin{align*}
rT_{tt} &= (f-2)((f-2)g' - (f-1)\dot{g})/g - ((f-2)f' - (f-1)\dot{f}) - (f-2)(f+g-2)/r - Λ g (f-2)r \\
rT_{tr} &= (f-2)((f-1)g' - f\dot{g})/g - (f-1)f' + f\dot{f} - (f-1)(f+g-2)/r - \Lambda g (f-1)r\\
rT_{rr} &= ((f^2 -2f +2)g' - f(f-1)\dot{g})/g  - ff' + (f+1)\dot{f} - f(f+g-2)/r - \Lambda fg r\\
\frac{-2g^2}{r^2}T_{\theta \theta} &=
-\frac{1}{g}(g'-\dot{g}) ((f-2)g' - f\dot{g}) + g \frac{2(f' - \dot{f})}{r} + g(f'' - 2 \dot{f}' + \ddot{f} )]] \\
& [(f-2)g''+(f'-\dot{f})(g'-\dot{g})- 2(f-1)\dot{g}'+f\ddot{g}] + 2g^2 \Lambda\\
\end{align*}

### We shall use the following equations:

The first is just a gauge condition, the second is the definition of the quantity $h$, which we shall use to render the system first order, in both space and time.

$$
\dot f - f' -\frac{f+g-2}{r} = 0
$$

$$
\dot g + \frac{g}{f} f' - \frac{f-2}{f} g' - 2gh + \frac{g}{f}\frac{f+g-2}{r} = 0
$$


From here we can get and equation for h, and then for its derivatives:

$$
\dot g - \frac{f-2}{f} g' - 2gh + \frac{g}{f}\dot f' = 0
$$

$$
h = \frac{1}{2g}(\dot g - \frac{f-2}{f} g') + \frac{1}{2f}\dot f
$$

$$
2gf\dot h = f\ddot g + g\ddot f - (f-2) \dot g' - \frac{f}{g}(\dot g - \frac{f-2}{f} g')\dot g - \frac{2}{f}g'\dot f -  \frac{g}{f}\dot f^2
$$

$$
2gfh' = f(\dot g'- \frac{f-2}{f} g'') + g\dot f' - \frac{f}{g}(\dot g - \frac{f-2}{f} g') g' - \frac{2}{f}g'f' -  \frac{g}{f}(\dot f f')
$$

We make the following linear combination of the equations: 

\begin{align}
& r[T_{tt}(f-1) - T_{rt}(f-2)] \\
&= (f-1)(f-2)^2 g'/g - (f-1)^2(f-2)\dot{g}/g - (f-1)(f-2)f' + (f-1)^2\dot{f} - (f-1)(f-2)(f+g -2)/r \\
& - (f-1)(f-2)^2g' + (f-1)(f-2) f' + f(f-2)^2\dot{g}/g - f(f-2)\dot{f} + (f-1)(f-2)(f+g-2)/r \\
&= \dot{g}/g(f-2)(-(f-1)^2 + f(f-2)) + \dot{f}((f-1)^2 - f(f-2)) \\
&= - \dot{g}/g(f-2) + \dot{f}
\end{align}

\begin{align}
& r[T_{tt}(f) - T_{rt}(f-1)] = \frac{2-f}{g} g' + f'+ \frac{f+g-2}{r} + \Lambda gr
\end{align}

\begin{align}
r[f\;T_{tr} - (f-1)T_{rr}] &= g'/g(f(f-2)(f-1) - (f-1)(f^2 -2f +2)) \\
&+ \dot{g}/g(-f^2(f-2) + f(f-1)^2)
+ f'(-f(f-1) + (f-1)f) + \dot{f}(f^2 - (f-1)(f+1)) \\
 &+ (f+g-2)/r(-f(f-1) + f(f-1)) - \Lambda gr(f(f-1) - f(f-1))\\
 &= -2(f-1)g'/g + f\dot{g}/g + \dot{f}
\end{align}

**So this is the $C_g$ constraint: (must check the energy-momentum side)**

$
r[f\;T_{tr} - (f-1)T_{rr}] = -2(f-1)g'/g + 2hf + (f-2)g'/g = 2hf -fg'/g
$

Or

$$
g'/g = 2h - r[T_{tr} - \frac{f-1}{f}T_{rr}] = 2h + \frac{rg}{f}T^t{}_r
$$

or 

$$
C_g = g' - 2hg - \frac{rg^2}{f}T^t{}_r
$$

**For the $C_f$ constraint:**

\begin{align}
& (2-f)r[f\;T_{tr} - (f-1)T_{rr}] + f r[T_{tt}(f) - T_{rt}(f-1)] = \\
&(2-f)(2hf -fg'/g) + f(\frac{2-f}{g} g' + f'+ \frac{f+g-2}{r} + \Lambda gr)\\
&= f[(2-f)2h + f' + \frac{f+g-2}{r} + \Lambda gr]
\end{align}

or 

\begin{align}
f' &= (f-2)2h - \frac{f+g-2}{r} + (2-f)r[T_{tr} - \frac{f-1}{f}T_{rr}] + r[T_{tt}(f) - T_{rt}(f-1)] - \Lambda gr \\
    &= (f-2)2h - \frac{f+g-2}{r} - \frac{gr(2-f)}{f}T^t{}_r - rg [T^t{}_t + \Lambda]
\end{align}

$$
C_f = f' - (f-2)2h + \frac{f+g-2}{r} + \frac{gr(2-f)}{f}T^t{}_r + rg [T^t{}_t + \Lambda]
$$

### We now go for the evolution equations: 

They can be deduced from the following equations and the constraints to elliminate the radial derivatives. The first equation is just a gauge condition, the second basically the definition of $h$.

$$
\partial_t f - \partial_r f -\frac{f+g-2}{r} = 0
$$

$$
\partial_t g + \frac{g}{f} \partial_r f - \frac{f-2}{f} \partial_r g - 2gh + \frac{g}{f}\frac{f+g-2}{r} = 0
$$


\begin{align}
\partial_t g &= - \frac{g}{f} \partial_r f + \frac{f-2}{f} \partial_r g + 2gh - \frac{g}{f}\frac{f+g-2}{r} \\
&= + 2gh - \frac{g}{f}\frac{f+g-2}{r} - \frac{g}{f}[(f-2)2h - \frac{f+g-2}{r} - \frac{gr(2-f)}{f}T^t{}_r - rg [T^t{}_t + \Lambda]] + \frac{f-2}{f}g[2h + \frac{rg}{f}T^t{}_r] \\
&= + 2gh   + \frac{g}{f} \frac{gr(2-f)}{f}T^t{}_r + \frac{g}{f} rg [T^t{}_t + \Lambda] + \frac{f-2}{f}g[\frac{rg}{f}T^t{}_r] \\ 
&= + 2gh   + \frac{g^2 r}{f^2}[[(2-f)  + (f-2)] T^t{}_r + f[T^t{}_t + \Lambda]] \\
&= + 2gh   + \frac{g^2 r}{f}[T^t{}_t + \Lambda]
\end{align}

### or

$$
\partial_t g = 2gh   + \frac{g^2 r}{f}[T^t{}_t + \Lambda]
$$

We now get the equation for $h$, which we obtain from $T_{θθ}$:

\begin{align*}
\frac{-2g^2}{r^2}T_{\theta \theta} &=-\frac{1}{g}(g'-\dot{g}) ((f-2)g'- f\dot{g}) + g \frac{2(f' - \dot{f})}{r} + g(f'' - 2 \dot{f}' + \ddot{f} ) \\
&+ (f-2)g''+(f'-\dot{f})(g'-\dot{g})- 2(f-1)\dot{g}'+f\ddot{g} + 2g^2 \Lambda\\
&= (g \ddot{f} + f\ddot{g}) -\frac{1}{g}(g'-\dot{g}) ((f-2)g'- f\dot{g}) + g \frac{2(f' - \dot{f})}{r} + g(f'' - 2 \dot{f}' ) \\
& (f-2)g''+(f'-\dot{f})(g'-\dot{g} - 2(f-1)\dot{g}' + 2g^2 \Lambda\\
&= 2gf \dot h - (- (f-2) \dot g' - \frac{f}{g}(\dot g - \frac{f-2}{f} g')\dot g - \frac{2}{f}g'\dot f -  \frac{g}{f}\dot f^2)) \\ 
&-\frac{1}{g}(g'-\dot{g}) ((f-2)g'- f\dot{g})  
+ g \frac{2(f' - \dot{f})}{r} + g(f'' - 2 \dot{f}' ) \\
& (f-2)g''+(f'-\dot{f})(g'-\dot{g})- 2(f-1)\dot{g}' + 2g^2 \Lambda\\
&= 2gf \dot h - (f\dot g'-(f-2)g'')\\
 &-(- (\frac{f}{g}(\dot g - \frac{f-2}{f} g')\dot g - \frac{2}{f}g'\dot f -  \frac{g}{f}\dot f^2)) -\frac{1}{g}(g'-\dot{g}) ((f-2)g'- f\dot{g}) + g \frac{2(f' - \dot{f})}{r} + g(f'' - 2 \dot{f}' ) \\
&+ (f'-\dot{f})(g'-\dot{g}) + 2g^2 \Lambda\\
&= 2gf \dot h - 2gfh' + g(f' - \dot{f})' + \frac{(f'-\dot f)}{f}((f- 2)g' - g\dot f  + \frac{2g}{r} - f\dot{g})\\ 
&= 2gf \dot h - 2gfh' + g(f' - \dot{f})' + \frac{(f'-\dot f)}{f}(-2gfh + \frac{2gf}{r})\\ 
&= 2gf \dot h - 2gfh' + \frac{1}{r}g(f'+g' - \frac{f+g-2}{r}) + \frac{f+g-2}{fr}(-2gfh + \frac{2gf}{r})\\
&= 2gf \dot h - 2gfh' \\ 
&+ \frac{1}{r}g(2h(f+g-2) - \frac{f+g-2}{r} -rg[T^t{}_t - \Lambda] + \frac{rg}{f}(f+g-2)T^t{}_r - \frac{f+g+2}{r})\\
 &+ \frac{f+g-2}{fr}(-2gfh + \frac{2gf}{r})\\
 &= 2gf (\dot h - h') -rg[T^t{}_t - \Lambda] + \frac{rg}{f}(f+g-2)T^t{}_r 
\end{align*} 

### Thus:

$$
\dot h = h' + \frac{r}{2f}([T^t{}_t - \Lambda] - \frac{(f+g-2)}{f}T^t{}_r) - 2g^2T^{\theta}{}_{\theta}

### The mass function is given by:

$$
m = \frac{r}{2}(1+\frac{f-2}{g})
$$

Thus, we have,

\begin{align}
\partial_t m &= \frac{r}{2}(\frac{\partial_t f}{g} - \frac{(f-2)\partial_t g}{g^2}) \\
&= \frac{r}{2}(\frac{(f-2)2h - \frac{gr(2-f)}{f}T^t{}_r - rg [T^t{}_t + \Lambda]}{g} - \frac{(f-2)(2gh + \frac{g^2r}{f} [T^t{}_t + Λ])}{g^2}) \\
&= \frac{r^2}{2}(- \frac{(2-f)}{f}T^t{}_r - [T^t{}_t + \Lambda] - (f-2)( \frac{1}{f} [T^t{}_t + Λ])) \\
&= \frac{r^2}{2f}(- (2-f)T^t{}_r - 2(f-1)[T^t{}_t + Λ]) \\
&= \frac{r^2g}{2f}(-g^{rr}[g^{tt}T_{tr} + g^{tr}T_{rr}] - 2g^{rt}[g^{tt}T_{tt} + g^{tr}T_{tr} + \Lambda])
\end{align}

In the case $0=g^{ab}T_{ab}= g^{tt}T_{tt} + 2g^{rt}T_{rt} + g^{rr}T_{rr}$ we can substitute $-g^{rr}T_{rr}$ and get,

\begin{align}
\partial_t m &=\frac{r^2g^2}{2f}(-g^{rr}[g^{tt}T_{tr}] + g^{tr}[g^{tt}T_{tt} + 2g^{rt}T_{rt}] - 2g^{rt}[g^{tt}T_{tt} + g^{tr}T_{tr} + \Lambda]) \\
&= \frac{r^2g}{2f}(-g^{rr}[g^{tt}T_{tr}] - g^{rt}[g^{tt}T_{tt}  + 2\Lambda]) \\
&= \frac{r^2g}{2f}[g^{tt}(-g^{rr} T_{tr} - g^{rt} T_{tt})  - g^{rt}2\Lambda] \\
&= \frac{-r^2g}{2f}[g^{tt}T^t{}_r + g^{rt}2\Lambda]
\end{align}
